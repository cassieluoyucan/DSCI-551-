```python
if __name__ == "__main__":
    if (len(sys.argv)!=4):# if user does not pass an input
        print ('Usage python filename csv_file_name')
        sys.exit(1)
        
    try:
        df_country = pd.read_json(sys.argv[1])  ### read data 
        
        ## filter GNP >10000
        df_1 = df_country[df_country["GNP"] >10000]
       
    ##  filter those continent less than 5 
        df_2 = df_1.groupby("Continent").filter(lambda g: len(g) >=5)
      
    
    ### find avg life expectancy for each continent 
        df_ans = df_2.groupby("Continent").LifeExpectancy.agg(["mean"])
   
        
        print(df_ans)
        
    
    except ValueError:
        print('error occur')
```
